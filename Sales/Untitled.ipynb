```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(123)
df = pd.DataFrame({"A1970" : {0 : "a", 1 : "b", 2 : "c"},
                   "A1980" : {0 : "d", 1 : "e", 2 : "f"},
                   "B1970" : {0 : 2.5, 1 : 1.2, 2 : .7},
                   "B1980" : {0 : 3.2, 1 : 1.3, 2 : .1},
                   "X"     : dict(zip(range(3), np.random.randn(3)))
                  })
df["id"] = df.index
df
```

```python
df['categories'] = 'A'
```

```python
df['categories'] = pd.Categorical(df['categories'], ['A'])
```

```python
df.categories = df.categories.astype('O')
```

```python
df.info()
```

```python
pd.wide_to_long(df, ["A", "B"], i='id', j="year").reset_index()
```
